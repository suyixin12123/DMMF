```python
# save 2011-2016 year movie id
info_file = open('../data/year2010_movie/movie_info.csv', 'w')
infor_file_w = csv.writer(info_file)
index = 0
movie_id_set = set()
with open('../data/ml-20m/movie_info.csv', encoding = "ISO-8859-1") as f:
    f_csv = csv.reader(f)
    headers = next(f_csv)
    print(headers)
    infor_file_w.writerow(headers)
    for row in f_csv:
        try:
            year = int(row[-1])
        except:
            year = 0
        if year > 2010:
            new_row = row
            new_row[0] = index
            index += 1
            infor_file_w.writerow(new_row)
            movie_id_set.add(row[1])

info_file.close()
"finish"
```

```python
headers = ['userid', 'movieid', 'rating', 'year']

year00 = open("../data/rating_time/movie_year_2011_2016.csv", 'w')
year00_f= csv.writer(year00)


year00_f.writerow(headers)


index = 0
droped = 0
with open('../data/ml-20m/ratings.csv') as f:
    f_csv = csv.reader(f)
    headers = next(f_csv)
    
    user_id = -1
    total_list = []
    for row in f_csv:
        index += 1
        #row = row[0].split("::")
        timestamp = int(row[3])
        year = time.localtime(timestamp).tm_year
        movieid = row[1]
        if movieid not in movie_id_set:
            droped += 1
            continue
        wrow = [row[0], row[1], row[2], year]
        
        if user_id != row[0]:
            if len(total_list) > 20:
                year00_f.writerows(total_list)
            
            total_list = [wrow]
            user_id = row[0]
        else:
            total_list.append(wrow)
            
        
                   
        if index%100000 == 0:
            print(index)
            
year00.close()
        
```
